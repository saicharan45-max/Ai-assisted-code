```python
def convert_temperature(value, from_unit, to_unit):
  """Converts temperature between Celsius, Fahrenheit, and Kelvin.

  Args:
    value: The temperature value to convert.
    from_unit: The unit to convert from ('C', 'F', or 'K').
    to_unit: The unit to convert to ('C', 'F', or 'K').

  Returns:
    The converted temperature value, or None if the units are invalid.
  """
  if from_unit == 'C':
    if to_unit == 'F':
      return (value * 9/5) + 32
    elif to_unit == 'K':
      return value + 273.15
    elif to_unit == 'C':
      return value
    else:
      return None
  elif from_unit == 'F':
    if to_unit == 'C':
      return (value - 32) * 5/9
    elif to_unit == 'K':
      return (value - 32) * 5/9 + 273.15
    elif to_unit == 'F':
      return value
    else:
      return None
  elif from_unit == 'K':
    if to_unit == 'C':
      return value - 273.15
    elif to_unit == 'F':
      return (value - 273.15) * 9/5 + 32
    elif to_unit == 'K':
      return value
    else:
      return None
  else:
    return None

# Example usage:
celsius_temp = 25
fahrenheit_temp = convert_temperature(celsius_temp, 'C', 'F')
kelvin_temp = convert_temperature(celsius_temp, 'C', 'K')

print(f"{celsius_temp}°C is {fahrenheit_temp:.2f}°F")
print(f"{celsius_temp}°C is {kelvin_temp:.2f}K")

fahrenheit_temp = 77
celsius_temp_from_f = convert_temperature(fahrenheit_temp, 'F', 'C')
kelvin_temp_from_f = convert_temperature(fahrenheit_temp, 'F', 'K')

print(f"{fahrenheit_temp}°F is {celsius_temp_from_f:.2f}°C")
print(f"{fahrenheit_temp}°F is {kelvin_temp_from_f:.2f}K")
```

```python
def convert_temperature_improved_error_handling(value, from_unit, to_unit):
  """Converts temperature between Celsius, Fahrenheit, and Kelvin with error handling.

  Args:
    value: The temperature value to convert.
    from_unit: The unit to convert from ('C', 'F', or 'K').
    to_unit: The unit to convert to ('C', 'F', or 'K').

  Returns:
    The converted temperature value, or None if the units or value are invalid.
  """
  valid_units = ['C', 'F', 'K']
  if from_unit not in valid_units or to_unit not in valid_units:
    print("Error: Invalid unit provided. Please use 'C', 'F', or 'K'.")
    return None

  # Attempt to convert the input value to a number
  try:
    value = float(value)
  except ValueError:
    print("Error: Invalid value provided. Please provide a number.")
    return None

  if from_unit == 'C':
    if to_unit == 'F':
      return (value * 9/5) + 32
    elif to_unit == 'K':
      return value + 273.15
    elif to_unit == 'C':
      return value
  elif from_unit == 'F':
    if to_unit == 'C':
      return (value - 32) * 5/9
    elif to_unit == 'K':
      return (value - 32) * 5/9 + 273.15
    elif to_unit == 'F':
      return value
  elif from_unit == 'K':
    if to_unit == 'C':
      return value - 273.15
    elif to_unit == 'F':
      return (value - 273.15) * 9/5 + 32
    elif to_unit == 'K':
      return value

# Get input from the user:
try:
  temp_value_str = input("Enter the temperature value: ")
  from_unit_input = input("Enter the unit to convert from (C, F, or K): ").upper()
  to_unit_input = input("Enter the unit to convert to (C, F, or K): ").upper()

  converted_temp = convert_temperature_improved_error_handling(temp_value_str, from_unit_input, to_unit_input)

  if converted_temp is not None:
    print(f"Converted temperature: {converted_temp:.2f}{to_unit_input}")

except Exception as e:
  print(f"An error occurred: {e}")
```
